# Eksploratorna analiza i priprema skupa podataka

```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
import scipy as sc
```

```python
# Popis kriptovaluti
coins = ['Aave', 'BinanceCoin', 'Bitcoin', 'Cardano', 'ChainLink','Cosmos', 'CryptocomCoin', 'Dogecoin', 'EOS', 'Ethereum', 'Iota', 'Litecoin', 'Monero', 'NEM', 'Polkadot', 'Solana', 'Stellar', 'Tether', 'Tron', 'USDCoin', 'Uniswap', 'WrappedBitcoin', 'XRP']
dataframes = {}

# U훾itavanje podataka
for coin in coins:
    file_path = f"../datasets/coin_{coin}.csv"
    try:
        dataframes[coin] = pd.read_csv(file_path)
    except FileNotFoundError:
        print(f"File for {coin} not found.")
        dataframes[coin] = None 
```

Pregled koli훾ine podataka po kriptovaluti.

```python
for coin,frame in dataframes.items():
    print(f"Number of values for {coin}: {len(frame)}")
```

```python
dataframes[coins[0]].head()
```

## Pregled metrika 

Ovo je korisno za provjeru to훾nosti puno kompleksnijih izra훾una u samom programu simulacije.

### Aave

```python
df = dataframes[coins[0]]
df['log_returns'] = np.log(df['Close'] / df['Open'])

plt.figure(figsize=(12, 7))
plt.plot(df.index, df['log_returns'], label='Vrijednost kroz vrijeme', color='deepskyblue', linewidth=2)
plt.title(coins[0] + ' povrati')
plt.xlabel('Datum')
plt.ylabel('Vrijednost')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 7))
plt.plot(df.index, df['Close'], label='Vrijednost kroz vrijeme', color='deepskyblue', linewidth=2)
plt.title(coins[0] + ' cijene')
plt.xlabel('Datum')
plt.ylabel('Vrijednost')
plt.legend()
plt.grid(True)
plt.show()
```

```python
mean_value = df['log_returns'].mean()
std_dev = df['log_returns'].std()

print(f"Prosje훾na vrijednost: {mean_value:.5f}")
print(f"Standardna devijacija: {std_dev:.5f}")
```
