# Eksploracijska analiza i priprema skupa podataka

```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
import scipy as sc
```

```python
# Popis kriptovaluti
coins = ['Aave', 'BinanceCoin', 'Bitcoin', 'Cardano', 'ChainLink','Cosmos', 'CryptocomCoin', 'Dogecoin', 'EOS', 'Ethereum', 'Iota', 'Litecoin', 'Monero', 'NEM', 'Polkadot', 'Solana', 'Stellar', 'Tether', 'Tron', 'USDCoin', 'Uniswap', 'WrappedBitcoin', 'XRP']
dataframes = {}

# Učitavanje podataka
for coin in coins:
    file_path = f"../datasets/coin_{coin}.csv"
    try:
        dataframes[coin] = pd.read_csv(file_path)
    except FileNotFoundError:
        print(f"File for {coin} not found.")
        dataframes[coin] = None 
```

Pregled količine podataka po kriptovaluti

```python
for coin,frame in dataframes.items():
    print(f"Number of values for {coin}: {len(frame)}")
```

```python
# Parametri simulacije
S0 = 100  # Početna cijena
mu = 0.05  # Očekivani godišnji prinos (5%)
sigma = 0.2  # Volatilnost (20%)
T = 1  # Vremenski period (1 godina)
N = 252  # Broj vremenskih koraka (252 trgovinska dana)
M = 10  # Broj simulacija

# Monte Carlo simulacija
dt = T / N
S = np.zeros((N, M))
S[0] = S0

for t in range(1, N):
    Z = np.random.standard_normal(M)
    print(S[t])
    S[t] = S[t-1] * np.exp((mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z)

# Vizualizacija simulacija
plt.figure(figsize=(10, 5))
plt.plot(S[:, :50], alpha=0.5)  # Prikazujemo samo 50 putanja radi čitljivosti
plt.xlabel("Dan")
plt.ylabel("Cijena")
plt.title("Monte Carlo simulacija kretanja cijene")
plt.show()
```
